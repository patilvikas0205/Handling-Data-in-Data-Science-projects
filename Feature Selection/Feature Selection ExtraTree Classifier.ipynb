```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("train.csv")  # mobile sales prediction
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
X=data.iloc[:,0:20]
X.head()
```

```python
Y=data.iloc[:,-1]
Y
```

```python
from sklearn.ensemble import ExtraTreesClassifier
select_feature=ExtraTreesClassifier()
select_feature.fit(X,Y)
```

```python
select_feature.feature_importances_
```

```python
feature_score=pd.DataFrame({"Data_Feature":X.columns,"Data_Score":select_feature.feature_importances_})
feature_score
```

```python
top_feature=feature_score.nlargest(10,"Data_Score")
top_feature
```

```python
top_feature_selction=list(top_feature["Data_Feature"])
top_feature_selction
```

```python
X=X[top_feature_selction]
```
