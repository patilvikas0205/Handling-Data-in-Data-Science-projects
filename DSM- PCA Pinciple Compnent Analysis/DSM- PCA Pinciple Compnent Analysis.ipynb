```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0",inplace=True,axis=1)
```

```python
df
```

```python
X=df[["TV","radio","newspaper"]]
```

```python
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
scaled_X=scalar.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
new_x=pca.fit_transform(scaled_X)
```

```python
new_x
```

```python
pca.explained_variance_
```

```python
pca.components_ #already Transpose
```

# by Mathematically

```python
center_x=scaled_X-scaled_X.mean()
```

```python
center_x
```

```python
center_xd=pd.DataFrame(center_x)
```

```python
center_xd
```

```python
mat=center_xd.cov()
```

```python
mat
```

```python
eig_value,eig_vector=np.linalg.eig(mat)
```

```python
eig_value
```

```python
eig_vector
```

```python
X_new=np.dot(center_xd,eig_vector[:2,:].T)
```
